```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Función cuadrática.
def f(x):
    return np.exp(x / 2) * np.sin(x / 3)
def g(x):
    return (1/3)*x
def h(x):
    return (1/3)*x + (1/6)*(x**2)
def j(x):
    return (1/3)*x + (1/6)*(x**2) +(23/648)*(x**3)
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace (-10, 10, 10)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(12, 8))
plt.title("Gráficas")
# Agregar rótulos a los ejes
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Agregar retícula
plt.grid(True)

plt.plot(x, f(x),label=r"$e^{x/2} \sin\left(\frac{x}{3}\right)$")

plt.plot(x, g(x),label=r"$\frac{1}{3}x$")

plt.plot(x,h(x), label=r"$\frac{1}{3}x +\frac{1}{6}x^2$")
plt.plot(x,j(x), label=r"$\frac{1}{3}x + \frac{1}{6}x^2 + \frac{23}{648}x^3$")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-30, 30)
plt.ylim(-30, 30)
plt.legend()
plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.show()
```
