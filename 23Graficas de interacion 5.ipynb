```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return (x**3) + 4 * (x**2) - 10
def g(x):
    return x-(x**3) - 4 * (x**2) + 10
def a(x):
    return ((10/x) - (4 * x))**(1/2)
def b(x):
    return (1/2) * (10 - x**3)**(1/2)
def d(x):
    return (10 / (4 + x))**(1/2)
def e(x):
    return x - ( (x**3 + 4*x**2 - 10) / (3*x**2 + 8*x) )
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(1, 5,100)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(12, 8))
plt.title("Gráficas")
# Agregar rótulos a los ejes
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Agregar retícula
plt.grid(True)

plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, a(x))
plt.plot(x, b(x))
plt.plot(x, d(x))
plt.plot(x, e(x))

plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(1, 2)
plt.ylim(1, 2)

plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.show()
```
