# Gr치fica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def g(x):
    return x

def h(x):
    return -3**(-x)

# Funci칩n cuadr치tica.
def f(x):
    return x - 3**(-x)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-5, 5,100)
```

```python
# Graficar ambas funciones
plt.figure(figsize=(12, 8))
plt.title("Gr치fica de $f(x)=x-3**(-x)$")
# Agregar r칩tulos a los ejes
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Agregar ret칤cula
plt.grid(True)


plt.plot(x,g(x))
plt.plot(x,h(x))
plt.plot(x,f(x))
plt.axhline(0, color="blue")
plt.axvline(0, color="black")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
plt.grid(color="pink")
# Mostrarlo.
plt.show()
```
