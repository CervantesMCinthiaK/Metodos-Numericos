# Gr치fica de serie de Taylor y polinomio dada por $f(x)=\sqrt{x+1}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Funci칩n cuadr치tica.
def f(x):
    return (x+1)**(1/2)
def g(x):
    return 1+(1/2)*x
def h(x):
    return (1+((1/2)*x)-(1/8)*(x**2))
def j(x):
    return (1+((1/2)*x)-((1/8)*(x**2))+((1/16)*(x**3)))
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-1, 10,100)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(12, 8))
plt.title("Gr치ficas")
# Agregar r칩tulos a los ejes
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Agregar ret칤cula
plt.grid(True)

plt.plot(x, f(x),label=r"$\sqrt{x+1}}$")

plt.plot(x, g(x),label=r"$1 + \frac{1}{2}x$")

plt.plot(x,h(x), label=r"$1 + \frac{1}{2}x - \frac{1}{8}x^2$")
plt.plot(x,j(x), label=r"$1 + \frac{1}{2}x - \frac{1}{8}x^2 + \frac{1}{16}x^3$")
#plt.axhline(0, color="black")
#plt.axvline(0, color="black")
plt.xlim(-1, 10)
plt.ylim(-1, 5)
plt.legend()
#plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
# Mostrarlo.
plt.show()
```
