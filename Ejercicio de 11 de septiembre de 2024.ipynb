#Gráfica de los Polinomios de Maclaurin dada la función $f(x)=e^{\frac{x}{2}}\sin\left(\frac{x}{3}\right)$


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import sympy as spy

x = spy.Symbol('x')
g=spy.exp(x/2)*spy.sin(x/3)
# Calcular la primera derivada
f_uno= spy.diff(g, x)
print(g)
print(f_uno)
```

```python
# Calcular la segunda derivada
f_dos= spy.diff(f_uno, x)
print(f_dos)
```

```python
#Calcular la tercera derivada
f_tres= spy.diff(f_dos, x)
print(f_tres)
```

```python
#Calcular la cuarta derivada
f_cuatro= spy.diff(f_tres, x)
print(f_cuatro)
```

```python
# Función cuadrática.
def f(x):
    return np.exp(x / 2) * np.sin(x / 3)
def g(x):
    return (1/3)*x
def h(x):
    return ((1/3)*x + (1/6)*(x**2))
def j(x):
    return (1/3)*x + (1/6)*(x**2) +(23/648)*(x**3)
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace (-10, 10, 400)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(10, 10))
plt.title("Gráficas de Polinomio de Maclaurin")
# Agregar rótulos a los ejes
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.axhline(0, color="grey")
plt.axvline(0, color="grey")
# Agregar retícula
plt.grid(True)

plt.plot(x, f(x),color="green",label=r"$f(x)=e^\frac{x}{2}\sin\left(\frac{x}{3}\right)$")

plt.plot(x, g(x),color="red",label=r"$P_{1}=\frac{1}{3}x$")

plt.plot(x,h(x), color="blue", label=r"$P_{2}=\frac{1}{3}x +\frac{1}{6}x^2$")

plt.plot(x,j(x), color="fuchsia",label=r"$P_{3}=\frac{1}{3}x + \frac{1}{6}x^2 + \frac{23}{648}x^3$")

plt.xlim(-10, 10)
plt.ylim(-20, 30)
plt.legend()
plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.show()
```

Esta gráfica resultó de hayar las derivadas de $f(x)=e^\frac{x}{2}\sin\left(\frac{x}{3}\right)$ , es decir: \( f \), \( f' \),\( f'' \) y
después evaluar cada una en x=0 y finalmente considerando cada resultado hayamos los polinomios: \\
$P_{1}=f(x_0) + f'(x_0) (x - x_0) = \frac{1}{3}x
$ \\
$P_{2}=f(x_0) + f'(x_0) (x - x_0) + \frac{f''(x_0)}{2!} (x - x_0)^2 = \frac{1}{3}x+\frac{1}{6}x^2  $
 \\
$P_{3}=f(x_0) + f'(x_0) (x - x_0) + \frac{f''(x_0)}{2!} (x - x_0)^2 + \frac{f'''(x_0)}{3!} (x - x_0)^3  =\frac{1}{3}x + \frac{1}{6}x^2 + \frac{23}{648}x^3$
